```python
# Create a text index on the 'title' and 'description' fields
movies_collection.create_index([('title', 'text'), ('fullplot', 'text')])
```

```python
# Perform a text search for the term 'adventure' and sort by relevance
search_results = movies_collection.find(
    {'$text': {'$search': 'adventure'}},
    {'score': {'$meta': 'textScore'}}
).sort([('score', {'$meta': 'textScore'})])
```

```python
# Display the search results
for movie in search_results:
    print(movie['title'], movie['score'])
```

```python
# Perform a text search for the term 'adventure' and filter by year greater than 2000
search_results = movies_collection.find(
    {
        '$text': {'$search': 'adventure'},
        'year': {'$gt': 2000}
    },
    {'score': {'$meta': 'textScore'}}
).sort([('score', {'$meta': 'textScore'})])

# Display the search results
for movie in search_results:
    print(movie['title'], movie['year'], movie['score'])
```

When performing a text search in MongoDB, the search results may include documents that match the search term in any of the indexed fields, not just the title. In this case, the term "adventure" might be found in the fullplot field instead of the title.

To ensure that the search term appears in the specific field you're interested in, you can use phrase search or additional filtering. Here’s how you can handle this:

Option 1: Phrase Search

Use double quotes to search for the exact phrase. This ensures that the term "adventure" is treated as a single search term:

```python
# Perform a text search for the exact phrase 'adventure' using the new index
search_results = movies_collection.find(
    {'$text': {'$search': '"adventure"'}},
    {'score': {'$meta': 'textScore'}}
).sort([('score', {'$meta': 'textScore'})])

# Display the search results
for movie in search_results:
    #print(movie['title'], movie['score'], movie.get('fullplot', 'No fullplot available'))
    print(movie['title'], movie['score'])
```

Option 2: Additional Filtering by Field

You can filter results to ensure that the search term appears in the title:

```python
# Perform a text search and filter results to ensure 'adventure' is in the 'title' field
search_results = movies_collection.find(
    {
        '$text': {'$search': 'adventure'},
        'title': {'$regex': 'adventure', '$options': 'i'}  # Case-insensitive regex search
    },
    {'score': {'$meta': 'textScore'}}
).sort([('score', {'$meta': 'textScore'})])

# Display the search results
for movie in search_results:
#     print(movie['title'], movie['score'], movie.get('fullplot', 'No fullplot available'))
     print(movie['title'], movie['score'])
```
